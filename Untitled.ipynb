```python
# X-y split
import pandas as pd
import numpy as np
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('files_for_lab/csv_files/marketing_customer_analysis.csv')
```

```python
data.head()
```

```python
data=data.drop(['Customer', 'Effective To Date'], axis=1)
```

```python
X=data.drop(['Total Claim Amount'], axis=1)
y=data['Total Claim Amount']
```

```python
X_num = X.select_dtypes(include = np.number)
X_cat = X.select_dtypes(include = object)
```

```python
X_num.describe()
```

```python
# Normalize numerical data using MinMaxScaler() to make data range from 0 - 1
transformer = MinMaxScaler().fit(X_num)
x_normalized = transformer.transform(X_num)
print(x_normalized.shape)
```

```python
pd.DataFrame(x_normalized, columns=X_num.columns)
```

```python
transformer = StandardScaler().fit(X_num)
x_standardized = transformer.transform(X_num)
print(x_standardized.shape)
pd.DataFrame(x_standardized, columns=X_num.columns)
```

```python
X_cat = data.select_dtypes(include = object)
X_cat
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder().fit(X_cat)
print(encoder.categories_)
encoder.categories_[0]
encoded = encoder.transform(X_cat).toarray()
print(encoded)
onehot_encoded = pd.DataFrame(encoded,columns=encoder.categories_)
onehot_encoded.head(20)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoded = LabelEncoder().fit(X_cat).transform(X_cat)
label_encoded = pd.DataFrame(label_encoded,columns=X_cat.columns)
display(label_encoded.head(20))
```
